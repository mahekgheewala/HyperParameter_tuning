GridSearchCV and RandomizedSearchCV

```python
# Importing Libraries
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

Converting/Loading this Data into pandas DataFrame

```python
data_frame = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
# Checking the Distribution of Target Variable
data_frame['label'].value_counts()
```

Seperating X and Y

```python
X = data_frame.drop(columns = 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

# GridSearchCV

It is used for determining the best parameters for our model

```python
# Loading SVC Model
model = SVC()
```

'kernel' : ['linear', 'poly', 'rbf', 'sigmoid']
You are telling the model:

“Try all 4 types of SVM kernels”

'linear' → Straight line boundary

'poly' → Polynomial curve

'rbf' → Radial basis (smooth curves)

'sigmoid' → Like neural networks (less common)

✅ 'C' : [1, 5, 10, 20]
C controls how much the model tries to fit the training data (called regularization).

You're telling it:

“Try C values of 1, 5, 10, and 20 and see which gives the best result.”

🔹 Smaller C = more general (less overfitting)
🔹 Bigger C = more strict on training accuracy (can overfit)

```python
# Hyperparameters

parameters = {
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'C' : [1,5,10,20]
}
```

 **Important:** for cv = 5, total kernel = 4 and C = 4, so there would be 4*4=16 combinations, and since cv = 5, hence each combination would be evaluated 5 times, so total 80 iterations. But it would show us the top 5 iterations with highest accuracy score.

```python
# Grid Search
classifier = GridSearchCV(model, parameters, cv=5)
# first we have to mention the model which we have to test
# then the parameters
# cv =  cross-validation (no. of folds)
```

```python
# Fitting the Data to our model
classifier.fit(X,Y)
```

```python
# cv_results---> prints us the result of the data
classifier.cv_results_
```

```python
# best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python

# higest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# Loading the results to Pandas DataFrame
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

mean_fit_time
Average time taken to train (fit) the model during each cross-validation run.

std_fit_time
How much the training time varies (standard deviation) across runs.

mean_score_time
Average time taken to test (score) the model during each run.

std_score_time
How much the scoring time varies across runs.

param_C
The value of the hyperparameter C used in the model (like SVM’s penalty strength).

param_kernel
The type of kernel used in the model (like linear, rbf, poly for SVM).

params
The full set of parameters/hyperparameters tried in that run (for example, C=1 and kernel='linear').

split0_test_score, split1_test_score, ..., split4_test_score
Accuracy (or other metric) scores of the model on each fold of the cross-validation (for each split of the data).

mean_test_score
Average accuracy (or metric) across all folds — the overall performance.

std_test_score
How much the test scores vary across the folds (standard deviation).

rank_test_score
The rank of the parameter set compared to others (1 means best score, 2 means second best, etc.).

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

# RandomizedSearchCV

```python
model = SVC()
```

```python
# hyperparamters
parameters = {
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1,5,10,20]
}
```

```python
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# higest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# Loading the results to Pandas DataFrame
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```
